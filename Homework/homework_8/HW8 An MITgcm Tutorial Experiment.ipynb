### Part 4: Assess the Model Results
Upon successful completion of your model run, you should have the following files: `THETA.0000070445.data`, `SALT.0000070445.data`, `UVEL.0000070445.data`, and `VVEL.0000070445.data`. These files contain the surface fields for temperature, salinity, zonal velocity, and meridional velocity in the month of December. 

Read these files into four separate grids and make a 2x2 set of subplots showing the surface temperature, salinity, zonal velocity, and meridional velocity simulated by the model. 

````python
# read in the data from files into grids for theta, salt, uvel, and vvel
theta_file = os.path.join(model_dir,'run','THETA.0000070445.data')
salt_file = os.path.join(model_dir,'run','SALT.0000070445.data')
uvel_file = os.path.join(model_dir,'run','UVEL.0000070445.data')
vvel_file = os.path.join(model_dir,'run','VVEL.0000070445.data')


theta_grid = np.fromfile(theta_file, '>f4').reshape((n_cols, n_rows))
salt_grid = np.fromfile(salt_file, '>f4').reshape((n_cols, n_rows))
uvel_grid = np.fromfile(uvel_file, '>f4').reshape((n_cols, n_rows))
vvel_grid = np.fromfile(vvel_file, '>f4').reshape((n_cols, n_rows))

X_file = os.path.join(model_dir,'run','XC.data') 
Y_file = os.path.join(model_dir,'run','YC.data')
X = np.fromfile(X_file, '>f4').reshape((n_cols, n_rows))
Y = np.fromfile(Y_file, '>f4').reshape((n_cols, n_rows))
````

````python
# make the 2x2 plot here
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
ax1 = axs[0, 0].pcolormesh(X, Y, theta_grid, cmap='turbo')
axs[0, 0].set_title('Surface Temperature')
fig.colorbar(ax1, ax=axs[0, 0], orientation='vertical')

ax2 = axs[0, 1].pcolormesh(X, Y, salt_grid, cmap='turbo')
axs[0, 1].set_title('Surface Salinity')
fig.colorbar(ax2, ax=axs[0, 1], orientation='vertical')

ax3 = axs[1, 0].pcolormesh(X, Y, uvel_grid, cmap='seismic', vmin=-0.2, vmax=0.2)
axs[1, 0].set_title('Zonal Velocity')
fig.colorbar(ax3, ax=axs[1, 0], orientation='vertical')

ax4 = axs[1, 1].pcolormesh(X, Y, vvel_grid, cmap='seismic', vmin=-0.2, vmax=0.2)
axs[1, 1].set_title('Meridional Velocity')
fig.colorbar(ax4, ax=axs[1, 1], orientation='vertical')
plt.tight_layout()
plt.show()
````
